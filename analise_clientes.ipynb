# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

```python
#Analisando 1 cliente
# Se o cliente é bom, ok ou ruim (Nota de Crédito = Score de Crédito)

import pandas as pd

tabela = pd.read_csv("data/clientes.csv")

display(tabela)
display(tabela.info())
```

```python
from sklearn.model_selection import train_test_split

# Colunas que serão usadas
colunas = [
    "idade",
    "salario_anual",
    "dias_atraso",
    "divida_total",
    "num_cartoes"
]

# Entrada da IA
x = tabela[colunas]

# Resultado esperado
y = tabela["score_credito"]

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)
```

```python
# passo 1: Importar a IA
from sklearn.ensemble import RandomForestClassifier
import joblib

# passo 2: Criar a IA
modelo_arvoreDecisao = RandomForestClassifier()

# passo 3: Treinar a IA
modelo_arvoreDecisao.fit(x_treino,y_treino)

#passo 4: Salvar o "cerebro" da IA
joblib.dump(modelo_arvoreDecisao, "modelo.pkl")
```

```python
previsao_arvore = modelo_arvoreDecisao.predict(x_teste)

from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore))
```

```python
novos_clientes = pd.read_csv("data/novos_clientes.csv")

previsao = modelo_arvoreDecisao.predict(novos_clientes[colunas])

display(novos_clientes)
display(previsao)

print(tabela["score_credito"].unique())
print(tabela["score_credito"].value_counts())
print(previsao)
```
